```python
from datetime import datetime
class politic():
    def __init__(self,name,startday,endday,pandasname):
        self.name = name
        self.startday = startday
        self.endday = endday
        self.pandasname = pandasname
    def transdate(self,selectday):
        self.selectday = selectday
        return(self.selectday)
    def select(self,startday,enday,wantday):
        self.starday = politic.transdate(self,startday)
        self.endday = politic.transdate(self,enday)
        self.wantday = politic.transdate(self,wantday)
        if self.startday < self.wantday < self.endday:
            return(self.pandasname[self.pandasname["page_name"]==self.name])
        else:
            return("不在範圍內")
    
#    def dayselect(self):
```
